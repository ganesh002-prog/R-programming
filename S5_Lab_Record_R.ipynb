Experiment 3: Functions in R

```r
# Q1: Bank Interest Calculator
cat("--- Q1: Simple Interest Calculator ---\n")
simple_interest <- function(p, r, t){
  si <- (p * r * t) / 100
  return(si)
}
print(simple_interest(10000, 5, 2))
```

```r
# Q2: Student Result Processing
cat("--- Q2: Student Result Check ---\n")
check_result <- function(marks){
  if(marks >= 40){ return("Pass") } else { return("Fail") }
}
print(check_result(65))
```

```r
# Q3: Employee Net Salary
cat("--- Q3: Net Salary After Tax ---\n")
net_salary <- function(basic_salary){
  tax <- basic_salary * 0.10
  net <- basic_salary - tax
  return(net)
}
print(net_salary(30000))
```

```r
# Q4: Electricity Bill Generator
cat("--- Q4: Electricity Bill ---\n")
electricity_bill <- function(units){
  if(units <= 100){
    bill <- units * 2
  } else if(units <= 200){
    bill <- (100 * 2) + (units - 100) * 3
  } else {
    bill <- (100 * 2) + (100 * 3) + (units - 200) * 5
  }
  return(bill)
}
print(electricity_bill(250))
```

```r
# Q5: Online Shopping Discount
cat("--- Q5: Final Payable Amount ---\n")
final_amount <- function(amount){
  if(amount >= 5000){ discount <- amount * 0.20
  } else if(amount >= 2000){ discount <- amount * 0.10
  } else { discount <- 0 }
  return(amount - discount)
}
print(final_amount(4500))
```

```r
# Q6: Temperature Conversion
cat("--- Q6: Celsius to Fahrenheit ---\n")
celsius_to_fahrenheit <- function(c){
  f <- (c * 9/5) + 32
  return(f)
}
print(celsius_to_fahrenheit(37))
```

```r
# Q7: Grading System
cat("--- Q7: Assign Grade ---\n")
assign_grade <- function(marks){
  if(marks >= 90){ "A"
  } else if(marks >= 75){ "B"
  } else if(marks >= 50){ "C"
  } else { "Fail" }
}
print(assign_grade(88))
```

```r
# Q8: Voting Eligibility Checker
cat("--- Q8: Voting Eligibility ---\n")
check_voting <- function(age){
  if(age >= 18){ return("Eligible to Vote")
  } else { return("Not Eligible to Vote") }
}
print(check_voting(17))
```

```r
# Q9: Bank Loan EMI Calculator
cat("--- Q9: EMI Calculator ---\n")
calculate_emi <- function(principal, rate, years){
  monthly_rate <- rate / (12 * 100)
  months <- years * 12
  emi <- (principal * monthly_rate * (1 + monthly_rate)^months) /
         ((1 + monthly_rate)^months - 1)
  return(round(emi, 2))
}
print(calculate_emi(500000, 7.5, 10))
```

```r
# Q10: Hospital Patient Billing
cat("--- Q10: Hospital Bill ---\n")
hospital_bill <- function(consultation, room_charge, days){
  subtotal <- consultation + (room_charge * days)
  tax <- subtotal * 0.05
  total <- subtotal + tax
  return(total)
}
print(hospital_bill(500, 2000, 3))
```

```r
# Q11: Online Shopping Order Total
cat("--- Q11: Order Total ---\n")
order_total <- function(cart_value){
  if(cart_value >= 3000){ discount <- cart_value * 0.10 } else { discount <- 0 }
  if(cart_value >= 5000){ delivery <- 0 } else { delivery <- 100 }
  return(cart_value - discount + delivery)
}
print(order_total(4200))
```

```r
# Q12: University GPA Classification
cat("--- Q12: GPA Classification ---\n")
calculate_gpa <- function(marks){
  avg <- mean(marks)
  if(avg >= 85){ result <- "Distinction"
  } else if(avg >= 70){ result <- "First Class"
  } else if(avg >= 50){ result <- "Second Class"
  } else { result <- "Fail" }
  return(result)
}
print(calculate_gpa(c(78, 85, 90, 72, 80)))
```

```r
# Q13: Smart Electricity Meter
cat("--- Q13: Smart Meter Bill ---\n")
smart_meter_bill <- function(peak_units, offpeak_units){
  bill <- (peak_units * 6) + (offpeak_units * 3)
  total_units <- peak_units + offpeak_units
  if(total_units > 300){ bill <- bill * 1.10 }
  return(bill)
}
print(smart_meter_bill(180, 150))
```

```r
# Q14: Insurance Premium
cat("--- Q14: Insurance Premium ---\n")
insurance_premium <- function(base, age, smoker){
  premium <- base
  if(age > 50){ premium <- premium * 1.20 }
  if(smoker){ premium <- premium * 1.30 }
  return(premium)
}
print(insurance_premium(5000, 55, TRUE))
```

```r
# Q15: Employee Payroll
cat("--- Q15: Net Salary (HRA + DA + Tax) ---\n")
payroll <- function(basic){
  hra <- basic * 0.20
  da <- basic * 0.10
  gross <- basic + hra + da
  tax <- gross * 0.12
  net <- gross - tax
  return(net)
}
print(payroll(40000))
```

```r
# Q16: Water Usage Classification
cat("--- Q16: Water Usage Status ---\n")
water_usage_status <- function(litres){
  if(litres <= 500){ "Normal"
  } else if(litres <= 1000){ "High Usage"
  } else { "Critical" }
}
print(water_usage_status(1200))
```

## 🔬 Experiment 7: Read CSV File & Analyze Data in R
> ⚠️ This experiment reads from a CSV file. The code below **creates a sample CSV first**, then analyzes it.

```r
# Step 0: Create sample CSV (since we don't have the file)
cat("--- Creating sample sales_data.csv ---\n")
sample_data <- data.frame(
  Date = c("2026-01-01","2026-01-01","2026-01-02","2026-01-02","2026-01-03","2026-01-03"),
  Product = c("Phone","Bag","Laptop","Headset","Bag","Phone"),
  Category = c("Electronics","Accessories","Electronics","Electronics","Accessories","Electronics"),
  Quantity = c(5,10,3,7,5,3),
  Price = c(20000,1500,50000,2500,1500,20000)
)
write.csv(sample_data, "sales_data.csv", row.names=FALSE)
cat("CSV created successfully!\n")
```

```r
# Step 1: Load CSV
sales <- read.csv("sales_data.csv")
head(sales); str(sales); summary(sales)
```

```r
# Step 2: Data Cleaning
sales$Product <- trimws(sales$Product)
sales$Category <- trimws(sales$Category)
sales$Date <- as.Date(sales$Date, format="%Y-%m-%d")
sales$TotalSales <- sales$Quantity * sales$Price
head(sales)
```

```r
# Step 3: Total Sales per Product
total_sales_product <- aggregate(TotalSales ~ Product, data=sales, sum)
print(total_sales_product)
```

```r
# Step 4: Top-Selling Product
top_product <- total_sales_product[which.max(total_sales_product$TotalSales), ]
print(top_product)
```

```r
# Step 5: Quantity per Category
quantity_per_category <- aggregate(Quantity ~ Category, data=sales, sum)
print(quantity_per_category)
```

```r
# Step 6: High-Value Products
high_sales <- total_sales_product[total_sales_product$TotalSales > 10000, ]
print(high_sales)
```

```r
# Step 7: Export
write.csv(total_sales_product, "total_sales_product.csv", row.names=FALSE)
cat("Exported to total_sales_product.csv\n")
```

## 🔬 Experiment 9: Statistical Analysis in R

```r
# Q1: School Student Marks Analysis
cat("--- Q1: Student Marks Statistical Analysis ---\n")
students <- data.frame(
  Name=c("Arjun","Meena","Rahul","Anitha","Kiran","Sneha","Ravi","Gayathri","Suresh","Lakshmi"),
  Maths=c(85,90,70,95,80,88,75,92,68,86),
  Science=c(78,88,60,92,85,80,70,90,75,82),
  English=c(92,80,75,98,88,85,78,91,80,87)
)
mean_marks <- sapply(students[,2:4], mean)
median_marks <- sapply(students[,2:4], median)
sd_marks <- sapply(students[,2:4], sd)
students$Total <- rowSums(students[,2:4])
top_student <- students[which.max(students$Total), ]
avg_math <- mean(students$Maths)
above_avg_math <- students[students$Maths > avg_math, ]
cat("Mean:\n"); print(mean_marks)
cat("Median:\n"); print(median_marks)
cat("SD:\n"); print(sd_marks)
cat("Top Student:\n"); print(top_student)
cat("Above Avg in Maths:\n"); print(above_avg_math)
```

```r
# Q2: Retail Sales Data Analysis
cat("--- Q2: Retail Sales Analysis ---\n")
sales <- data.frame(
  Product=c("Phone","Laptop","Headset","Bag","Camera","Mouse","Keyboard"),
  Day1=c(5,3,7,10,2,8,6), Day2=c(4,2,8,12,3,7,5),
  Day3=c(6,4,6,8,1,9,7), Day4=c(5,3,7,11,2,6,6), Day5=c(7,5,5,9,3,8,8)
)
sales_stats <- data.frame(
  Product=sales$Product,
  Mean=apply(sales[,2:6], 1, mean),
  Max=apply(sales[,2:6], 1, max),
  Min=apply(sales[,2:6], 1, min),
  Variance=apply(sales[,2:6], 1, var)
)
max_avg_product <- sales_stats[which.max(sales_stats$Mean), ]
print(sales_stats)
cat("Max Avg Product:\n"); print(max_avg_product)
```

```r
# Q3: Hospital Patient Age Analysis
cat("--- Q3: Patient Age Analysis ---\n")
patients <- data.frame(
  PatientID=201:210,
  Name=c("Suresh","Lakshmi","Arjun","Meena","Ravi","Anitha","Kiran","Sneha","Gayathri","Ravi2"),
  Age=c(45,38,52,29,60,48,55,35,42,50)
)
cat("Mean Age:", mean(patients$Age), "\n")
cat("Median Age:", median(patients$Age), "\n")
cat("Range:", range(patients$Age), "\n")
cat("SD:", sd(patients$Age), "\n")
cat("Patients > 50:\n")
print(patients[patients$Age > 50, ])
```

```r
# Q4: Employee Salary Analysis
cat("--- Q4: Salary Analysis ---\n")
employees <- data.frame(
  EmpID=101:110,
  Name=c("Rahul","Sneha","Kiran","Anitha","Ravi","Gayathri","Suresh","Lakshmi","Arjun","Meena"),
  Dept=c("IT","HR","Finance","IT","HR","Finance","IT","HR","Finance","IT"),
  Salary=c(30000,35000,40000,32000,36000,42000,31000,37000,45000,33000)
)
dept_stats <- aggregate(Salary ~ Dept, data=employees, function(x) c(Mean=mean(x), Median=median(x)))
dept_stats <- do.call(data.frame, dept_stats)
overall_avg <- mean(employees$Salary)
high_earners <- employees[employees$Salary > overall_avg, ]
print(dept_stats)
cat("High Earners (above avg):\n"); print(high_earners)
```

```r
# Q5: Customer Ratings Analysis
cat("--- Q5: Product Ratings Analysis ---\n")
ratings <- data.frame(
  ProductID=1:10,
  Product=c("Phone","Laptop","Headset","Bag","Camera","Mouse","Keyboard","Charger","Tablet","Speaker"),
  Rating=c(4,5,3,4,5,4,3,5,4,4)
)
getmode <- function(v){ uniqv <- unique(v); uniqv[which.max(tabulate(match(v, uniqv)))] }
cat("Mean Rating:", mean(ratings$Rating), "\n")
cat("Median Rating:", median(ratings$Rating), "\n")
cat("Mode Rating:", getmode(ratings$Rating), "\n")
cat("Variance:", var(ratings$Rating), "\n")
cat("Top Rated (> 4):\n")
print(ratings[ratings$Rating > 4, ])
```
